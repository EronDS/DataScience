#Introduction

The **Selfish Herd Theory** was proposed by W.D.Hamilton and behaviours that follow the description given are widely spread in nature. 



In Selfish Herd behaviour, the individuals of a species are accompanied by several other conspecific individuals, thus diminishing the individual probability of being caught and enhancing their own direct fitness(the herd behaviour can elicit extra responses in the predator such as confusion, fear elicited by herding faking a large size animal, increment in the amplitude of mechanical waves by a phenomenon known as  **wave interference**, etc).





The selfish herd is an example of how a selfish-gene eyes view of evolution (Individual Selection) can select for cooperation between animals.

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import plotly.express as px 
import matplotlib.pyplot as plt 
```

```python
ProbabilityOfBeingPredated = .75 
```

```python
def DilutionEffect(ProbabilityOfBeingPredated = .75, groupsize = 10):
  p = []
  label = []
  generation = [] 
  P1 = ProbabilityOfBeingPredated
  P2 = ProbabilityOfBeingPredated - .25
  P3 = ProbabilityOfBeingPredated - .50
  count = 1
  for i in range(groupsize):
    z = 1 + i
    p.append(P1 / z)
    label.append('PopulationI p(i) = {}'.format(P1))
    generation.append(count)
    p.append(P2 / z)
    label.append('PopulationII p(i) = {}'.format(P2))
    generation.append(count)
    p.append(P3 / z)
    label.append('PopulationIII p(i) = {}'.format(P3))
    generation.append(count)
    count += 1

  
  return (p, label, generation)

```

```python
p, label, generation = DilutionEffect(ProbabilityOfBeingPredated = .75 , groupsize = 40)
```

```python
df = pd.DataFrame()
df['p'] = p 
df['Population'] = label
df['generations'] = generation
```

```python
plt.figure(figsize = (12,8))
sns.set_style('whitegrid')
sns.lineplot(x = df.generations , y = df.p , hue = df.Population , palette = 'Pastel1', 
             linewidth = 6, ).set(xlim = (1, 40))

plt.title('Probability of an Individual within the Herd being Predated',
          fontsize = 14 , loc = 'left', fontweight = 'bold')
plt.ylabel('Dilution Effect', fontsize  = 11, )
plt.xlabel('No of Individuals within the Herd', fontsize = 12 , fontweight = 'bold')
```
