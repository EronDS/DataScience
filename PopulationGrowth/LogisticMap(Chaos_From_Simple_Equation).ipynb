```python
import numpy as np 
import matplotlib.pyplot as plt

from scipy.integrate import odeint
import pandas as pd


plt.style.use('bmh')
plt.style.use('dark_background')
```

```python
plt.style.use('dark_background')

y0 = [1] #try w .5 #initial condition of x

t = np.linspace(0, 500 , num = 1000) #timesteps


r = 0.05 #growth rate

k = 100000 #carrying capacity 

parameters = [r , k]



def Growth(variables,t, parameters):

  x = variables[0]

  r = parameters[0]
  k = parameters[1]

  dxdt = (r * x )* ( 1 - x / k )

  return ([dxdt])



y = odeint(Growth, y0 , t , args = (parameters,))

plt.figure(figsize = (20,10))
plt.plot(t, y[:,0], linewidth = 5)
plt.xlabel('Time(t)')
plt.ylabel('Population(N)', fontsize = 14)
plt.show()
```
